## 3.2 Data Type Correction 

### 3.2.1 Change Feature Data Type To Numeric

In the dataset, the features price, security_deposit, cleaning_fee, extra_people and host_response_rate are in string format. It is necessary to convert their data type to numeric. 

```python
col_to_numeric = ['price','security_deposit','cleaning_fee','extra_people']
for col in col_to_numeric:
    df[col] = df[col].str.replace('$','').str.replace(',','').astype(float)
df.host_response_rate = df.host_response_rate.str.replace('%','').astype(float)/100
```

### 3.2.2 Change Feature Data Type to Datetime

In the dataset, the features last_scraped,host_since,calendar_last_scraped,first_review and last_review are in string format. It is necessary to convert their data type to numeric.

```python
col_to_datetime = ['last_scraped','host_since','calendar_last_scraped','first_review','last_review']
for col in col_to_datetime:
    df[col] = pd.to_datetime(df[col])
```

## 3.3 Missing Values Imputation

For this section, we will only focus on features that will likely be relevent to data analysis.  
For imputation of each feature, we will create a new column with name 'variable_NA' to identify the rows that originally consists of the missing value. It may be useful information if the reason for missing is systemic (i.e. non-random).  

### 3.3.1 Numeric Features

Various numeric features consist of missing values. They will be grouped by the method of imputation.  
For the features host_listings_count, host response rate, bathrooms, bedrooms, beds, the missing values will be imputed with the feature median values.

```python
for col in ['host_listings_count','bathrooms','host_response_rate','bedrooms','beds']:
    df[col+'_NA'] = np.where(df[col].isnull(),True, False)
    m = df[col].median()
    df[col].fillna(m,inplace=True)
```

Missing security deposits and cleaning fees are likely due the hosts' decision to not include one, which is equivalent to a value of 0. As such, the missing values will be imputed with 0.

```python
for col in ['security_deposit','cleaning_fee']:
    df[col+'_NA'] = np.where(df[col].isnull(),True, False)
    df[col].fillna(0,inplace=True)
```

For review scores, the missing values are likely due to the facts that either the listings are new with few customers, or their customers did not leave a review score. The missing values are imputed with the feature median values.

```python
for col in ['review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin',
           'review_scores_communication','review_scores_location','review_scores_value','reviews_per_month']:
    df[col+'_NA'] = np.where(df[col].isnull(),True, False)
    df[col].fillna(df[col].median(),inplace=True)
```

### 3.3.3 Datetime Features

The datetime features with missing values are host_since, first_review, and last_review. The missing values are imputed with feature medians.

```python
for col in ['host_since', 'first_review', 'last_review']:
    df[col+"_NA"] = np.where(df[col].isnull(),1,0)
    df[col].fillna(df[col].quantile(0.5),inplace=True)
```

## 3.4 New Feature Creation 

### 3.4.1 Dummy Variable for Amenities

The feature amenities contains a list of attributes provided by the host that the property contains. It is stored as a string enclosed in curly brackets with each attribute separated by comma. To further evaluate those amenities, a dummy variable is created for each amenity, with '1' and '0' indicating the presence and absence of that amenity, respectively. 

```python
df['amenities_list'] = df['amenities'].str.lower().str.replace('[{}"]','').str.split(',')
amenities_set = set()
for i in df['amenities_list']:
    amenities_set.update(i)
amenities_set.remove('')
for item in amenities_set:
    df[item] = df.amenities_list.apply(lambda x: 1 if item in x else 0) 
```

There are 196 unique amenities.

## 3.4.2 City Names

One way to identify the geographical location of each listing is to map the city they are located in through its zip code. The first 3 characters of Canada's postal code, known as the Forward Sortation Area (FSA), will be used.  

```python
df['fsa'] = df.zipcode.str.slice(stop=3).str.upper()
toronto_fsa = pd.read_csv('data/toronto_fsa.csv')
fsa_dict = {i:j for i,j in zip(toronto_fsa.Postcode, toronto_fsa.Borough)}
df['city_fsa'] = np.nan
df['city_fsa'] = df.fsa.map(fsa_dict)
```
